```
from textblob import TextBlob
def nltk_tblob(sentence):
  p=dict()
  blob = TextBlob(sentence)
  for sentence in blob.sentences:
     p["Polarity"] =sentence.sentiment.polarity
  return p

df['TweetText_blob']= df['TweetText'].apply(nltk_tblob)
print(df["TweetText_blob"])

df1['reviewText_blob']= df1['reviewText'].apply(nltk_tblob)
print(df1["reviewText_blob"])
```

```
from textblob import TextBlob
def nltk_tblob(sentence):
  p=dict()
  blob = TextBlob(sentence)
  for sentence in blob.sentences:
     return sentence.sentiment.polarity

def classify(pol):
  if pol>0:
    return "positive"
  elif pol<0:
    return "negetive"
  else:
    return "neutral"

df['TweetText_blob']= df['TweetText'].apply(nltk_tblob)
#print(df["TweetText_blob"])

df['TweetText_blob1']= df['TweetText_blob'].apply(classify)
print(df["TweetText_blob1"])
```

```
from glove import Corpus,Glove
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger.pickle')
nltk.download('all')
Tweets = df["TweetText"].apply(nltk.word_tokenize)
corpus = Corpus()
corpus.fit(Tweets,window=10)
glove = Glove(no_components=100, learning_rate=0.05)
glove.fit(corpus.matrix, epochs=30, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)
glove.most_similar('amazing')
```
